```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n de parÃ¡metros optimizados
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 16
EPOCHS_STAGE1 = 8
EPOCHS_STAGE2 = 6

class DogCatClassifier:
    def __init__(self):
        self.data_path = self.setup_google_drive()
        self.model = None
        self.class_names = ['cats', 'dogs']
        self.use_transfer_learning = True

    def setup_google_drive(self):
        """Configura la conexiÃ³n con Google Drive"""
        try:
            from google.colab import drive
            drive.mount('/content/drive')
            print("âœ… Google Drive montado correctamente")

            folder_id = "1IH-4TdPIe_xxMjyt__LTpD7rvUFsNplC"
            print("ğŸ”— Accediendo a tu carpeta especÃ­fica de Google Drive...")

            direct_path = f"/content/drive/MyDrive/.shortcut-targets-by-id/{folder_id}"

            possible_paths = [
                direct_path,
                f"/content/drive/MyDrive/Data_Set",
                f"/content/drive/MyDrive/.shortcut-targets-by-id/{folder_id}/Data_Set",
                f"/content/drive/MyDrive/Colab Notebooks/Data_Set",
                "/content/drive/MyDrive/dataset",
                "/content/Data_Set"
            ]

            for path in possible_paths:
                if os.path.exists(path):
                    if self.verify_dataset_structure(path):
                        print(f"ğŸ“ Dataset encontrado en: {path}")
                        return path

            print("ğŸ” Buscando dataset recursivamente en Google Drive...")
            found_path = self.find_dataset_recursive("/content/drive/MyDrive", folder_id)
            if found_path:
                return found_path

            print("ğŸ” BÃºsqueda final en todo Google Drive...")
            for root, dirs, files in os.walk("/content/drive/MyDrive"):
                if "Data_Set" in dirs:
                    potential_path = os.path.join(root, "Data_Set")
                    if self.verify_dataset_structure(potential_path):
                        print(f"ğŸ¯ Dataset encontrado en: {potential_path}")
                        return potential_path

            return self.create_minimal_dataset()

        except Exception as e:
            print(f"âš ï¸ Error con Google Drive: {e}")
            return self.find_local_path()

    def find_dataset_recursive(self, start_path, folder_id):
        """Busca recursivamente la carpeta especÃ­fica"""
        shortcut_path = f"/content/drive/MyDrive/.shortcut-targets-by-id/{folder_id}"
        if os.path.exists(shortcut_path):
            if self.verify_dataset_structure(shortcut_path):
                return shortcut_path
            else:
                for item in os.listdir(shortcut_path):
                    item_path = os.path.join(shortcut_path, item)
                    if os.path.isdir(item_path) and self.verify_dataset_structure(item_path):
                        return item_path

        for root, dirs, files in os.walk(start_path):
            if 'cats' in dirs and 'dogs' in dirs:
                if self.verify_dataset_structure(root):
                    print(f"ğŸ¯ Dataset encontrado en: {root}")
                    return root
            if folder_id in root:
                if self.verify_dataset_structure(root):
                    return root
        return None

    def verify_dataset_structure(self, path):
        """Verifica que la carpeta tenga la estructura correcta"""
        try:
            cats_path = os.path.join(path, 'cats')
            dogs_path = os.path.join(path, 'dogs')

            if os.path.exists(cats_path) and os.path.exists(dogs_path):
                cat_images = [f for f in os.listdir(cats_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
                dog_images = [f for f in os.listdir(dogs_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

                if cat_images and dog_images:
                    print(f"âœ… Estructura vÃ¡lida en {path}: {len(cat_images)} gatos, {len(dog_images)} perros")

                    if len(cat_images) < 100 or len(dog_images) < 100:
                        print("âš ï¸  Dataset pequeÃ±o, considera aÃ±adir mÃ¡s imÃ¡genes")

                    return True
                else:
                    print(f"âš ï¸  Carpeta encontrada pero sin imÃ¡genes: {path}")
            return False
        except Exception as e:
            print(f"âŒ Error verificando {path}: {e}")
            return False

    def create_minimal_dataset(self):
        """Crea un dataset mÃ­nimo si no se encuentra el original"""
        print("ğŸ“ Creando dataset mÃ­nimo de demostraciÃ³n...")

        minimal_path = "/content/minimal_dataset"
        os.makedirs(os.path.join(minimal_path, "cats"), exist_ok=True)
        os.makedirs(os.path.join(minimal_path, "dogs"), exist_ok=True)

        print("\nâŒ NO SE ENCONTRÃ“ EL DATASET EN TU GOOGLE DRIVE")
        print("ğŸ’¡ POR FAVOR VERIFICA:")
        print("1. âœ… Que hayas montado Google Drive correctamente")
        print("2. âœ… Que la carpeta 'Data_Set' exista en tu Drive")
        print("3. âœ… Que dentro de 'Data_Set' estÃ©n las carpetas 'cats' y 'dogs'")
        print("4. âœ… Que las carpetas tengan imÃ¡genes JPG/PNG")
        print(f"5. âœ… Que el enlace sea correcto: https://drive.google.com/drive/folders/1IH-4TdPIe_xxMjyt__LTpD7rvUFsNplC")

        input("\nâ³ Presiona Enter despuÃ©s de verificar tu Google Drive...")

        print("ğŸ”„ Reintentando bÃºsqueda...")
        return self.setup_google_drive()

    def find_local_path(self):
        """Para uso local"""
        local_paths = ["Data_Set", "./Data_Set", "../Data_Set", "dataset"]
        for path in local_paths:
            if os.path.exists(path) and self.verify_dataset_structure(path):
                return path
        return "Data_Set"

    def create_advanced_model(self):
        """Modelo CNN avanzado con mejores tÃ©cnicas"""
        if self.use_transfer_learning:
            return self.create_transfer_learning_model()
        else:
            return self.create_custom_cnn_model()

    def create_transfer_learning_model(self):
        """Modelo con Transfer Learning para mejor precisiÃ³n"""
        print("ğŸ§  Creando modelo con Transfer Learning (MobileNetV2)...")

        base_model = tf.keras.applications.MobileNetV2(
            input_shape=(IMG_HEIGHT, IMG_WIDTH, 3),
            include_top=False,
            weights='imagenet'
        )

        base_model.trainable = False

        model = tf.keras.Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dropout(0.3),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.2),
            layers.Dense(128, activation='relu'),
            layers.Dropout(0.1),
            layers.Dense(1, activation='sigmoid')
        ])

        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
            loss='binary_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        return model

    def create_custom_cnn_model(self):
        """CNN personalizado mejorado"""
        print("ğŸ§  Creando CNN personalizado avanzado...")

        model = keras.Sequential([
            layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.2),

            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.25),

            layers.Conv2D(128, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.3),

            layers.Conv2D(256, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2, 2),
            layers.Dropout(0.4),

            layers.Flatten(),
            layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(256, activation='relu'),
            layers.Dropout(0.3),
            layers.Dense(1, activation='sigmoid')
        ])

        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
            loss='binary_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        return model

    def unfreeze_layers(self):
        """Descongela capas para fine-tuning"""
        if self.use_transfer_learning and self.model is not None:
            base_model = self.model.layers[0]
            base_model.trainable = True

            for layer in base_model.layers[:-20]:
                layer.trainable = False
            for layer in base_model.layers[-20:]:
                layer.trainable = True

            print("âœ… Capas descongeladas para fine-tuning")

    def prepare_advanced_data(self):
        """Prepara los datos con data augmentation mejorado"""
        print("ğŸ“Š Preparando datos con augmentaciÃ³n avanzada...")

        if not os.path.exists(self.data_path):
            print(f"âŒ La ruta {self.data_path} no existe")
            return None, None, None

        train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
            rescale=1./255,
            rotation_range=12,
            width_shift_range=0.08,
            height_shift_range=0.08,
            shear_range=0.08,
            zoom_range=0.12,
            horizontal_flip=True,
            brightness_range=[0.92, 1.08],
            channel_shift_range=0.05,
            fill_mode='nearest',
            validation_split=0.2
        )

        val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
            rescale=1./255,
            validation_split=0.2
        )

        try:
            train_generator = train_datagen.flow_from_directory(
                self.data_path,
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                subset='training',
                shuffle=True
            )

            validation_generator = val_datagen.flow_from_directory(
                self.data_path,
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                subset='validation',
                shuffle=False
            )

            print(f"âœ… {train_generator.samples} imÃ¡genes de entrenamiento")
            print(f"âœ… {validation_generator.samples} imÃ¡genes de validaciÃ³n")

            class_weights = self.calculate_class_weights(train_generator)
            print(f"âš–ï¸  Pesos de clases: {class_weights}")

            return train_generator, validation_generator, class_weights

        except Exception as e:
            print(f"âŒ Error preparando datos: {e}")
            return None, None, None

    def calculate_class_weights(self, generator):
        """Calcula pesos para balancear clases"""
        from sklearn.utils.class_weight import compute_class_weight

        classes = generator.classes
        class_weights = compute_class_weight(
            'balanced',
            classes=np.unique(classes),
            y=classes
        )

        return {i: weight for i, weight in enumerate(class_weights)}

    def train_advanced(self):
        """Entrenamiento avanzado en dos etapas"""
        try:
            print("ğŸš€ Iniciando entrenamiento avanzado...")
            train_generator, validation_generator, class_weights = self.prepare_advanced_data()

            if train_generator is None:
                print("âŒ No se pudieron cargar los datos de entrenamiento")
                return None

            self.model = self.create_advanced_model()

            print("ğŸ“‹ Resumen del modelo:")
            self.model.summary()

            callbacks = [
                tf.keras.callbacks.EarlyStopping(
                    monitor='val_accuracy',
                    patience=6,
                    restore_best_weights=True,
                    verbose=1
                ),
                tf.keras.callbacks.ReduceLROnPlateau(
                    monitor='val_loss',
                    factor=0.3,
                    patience=4,
                    min_lr=0.00001,
                    verbose=1
                ),
                tf.keras.callbacks.ModelCheckpoint(
                    'best_dog_cat_model_advanced.h5',
                    monitor='val_accuracy',
                    save_best_only=True,
                    save_weights_only=False,
                    verbose=1
                ),
                tf.keras.callbacks.CSVLogger('training_log.csv')
            ]

            print(f"\nğŸ¯ ETAPA 1: Entrenamiento bÃ¡sico ({EPOCHS_STAGE1} Ã©pocas)")
            history1 = self.model.fit(
                train_generator,
                steps_per_epoch=max(1, train_generator.samples // BATCH_SIZE),
                epochs=EPOCHS_STAGE1,
                validation_data=validation_generator,
                validation_steps=max(1, validation_generator.samples // BATCH_SIZE),
                callbacks=callbacks,
                class_weight=class_weights,
                verbose=1
            )

            if self.use_transfer_learning:
                print(f"\nğŸ¯ ETAPA 2: Fine-tuning ({EPOCHS_STAGE2} Ã©pocas)")
                self.unfreeze_layers()

                self.model.compile(
                    optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001),
                    loss='binary_crossentropy',
                    metrics=['accuracy', 'precision', 'recall']
                )

                history2 = self.model.fit(
                    train_generator,
                    steps_per_epoch=max(1, train_generator.samples // BATCH_SIZE),
                    epochs=EPOCHS_STAGE2,
                    validation_data=validation_generator,
                    validation_steps=max(1, validation_generator.samples // BATCH_SIZE),
                    callbacks=callbacks,
                    class_weight=class_weights,
                    verbose=1
                )

                history = self.combine_histories(history1, history2)
            else:
                history = history1

            if os.path.exists('best_dog_cat_model_advanced.h5'):
                self.model = tf.keras.models.load_model('best_dog_cat_model_advanced.h5')
                print("âœ… Mejor modelo cargado")

            self.model.save('dog_cat_classifier_advanced.h5')
            print("ğŸ’¾ Modelo final guardado")

            self.evaluate_model(validation_generator, history)

            return history

        except Exception as e:
            print(f"âŒ Error durante el entrenamiento: {e}")
            import traceback
            traceback.print_exc()
            return None

    def combine_histories(self, history1, history2):
        """Combina los historiales de las dos etapas"""
        combined_history = {}
        for key in history1.history.keys():
            combined_history[key] = history1.history[key] + history2.history[key]
        return type('History', (), {'history': combined_history})

    # ğŸ†• NUEVO MÃ‰TODO PARA CUMPLIR REQUERIMIENTOS
    def evaluate_classification_metrics(self, test_generator=None):
        """ğŸ“Š CALCULA TODAS LAS MÃ‰TRICAS REQUERIDAS FORMALMENTE"""
        print("\n" + "="*70)
        print("ğŸ¯ EVALUACIÃ“N FORMAL - MÃ‰TRICAS DE CLASIFICACIÃ“N")
        print("="*70)

        if self.model is None:
            print("âŒ No hay modelo entrenado. Entrena primero el modelo.")
            return None

        if test_generator is None:
            _, test_generator, _ = self.prepare_advanced_data()
            if test_generator is None:
                print("âŒ No se pudo cargar el conjunto de prueba")
                return None

        try:
            # ğŸ¯ REQUERIMIENTO 1: Usar .predict() sobre el conjunto de prueba
            print("ğŸ”® Realizando predicciones con .predict(X_test)...")
            y_pred_proba = self.model.predict(test_generator, verbose=1)
            y_pred = (y_pred_proba > 0.5).astype(int).flatten()
            y_true = test_generator.classes

            # ğŸ¯ REQUERIMIENTO 2: Calcular Accuracy
            from sklearn.metrics import accuracy_score
            accuracy = accuracy_score(y_true, y_pred)
            print(f"\nâœ… PRECISIÃ“N (ACCURACY): {accuracy:.4f} ({accuracy*100:.2f}%)")

            # ğŸ¯ REQUERIMIENTO 3: Matriz de ConfusiÃ³n
            from sklearn.metrics import confusion_matrix
            import seaborn as sns

            cm = confusion_matrix(y_true, y_pred)
            print(f"\nğŸ“‹ MATRIZ DE CONFUSIÃ“N:")
            print(f"[[Verdaderos Gatos    Falsos Perros]")
            print(f" [Falsos Gatos      Verdaderos Perros]]")
            print(f"\n{cm}")

            # VisualizaciÃ³n de la matriz de confusiÃ³n
            plt.figure(figsize=(8, 6))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                       xticklabels=['Gato', 'Perro'],
                       yticklabels=['Gato', 'Perro'])
            plt.title('Matriz de ConfusiÃ³n', fontsize=16, fontweight='bold')
            plt.ylabel('Etiqueta Real')
            plt.xlabel('Etiqueta Predicha')
            plt.show()

            # ğŸ¯ REQUERIMIENTO 4: PrecisiÃ³n (Precision)
            from sklearn.metrics import precision_score
            precision = precision_score(y_true, y_pred)
            print(f"\nğŸ¯ PRECISIÃ“N (PRECISION): {precision:.4f}")
            print("   - Calidad de las predicciones positivas (Perros)")

            # ğŸ¯ REQUERIMIENTO 5: Exhaustividad (Recall)
            from sklearn.metrics import recall_score
            recall = recall_score(y_true, y_pred)
            print(f"\nğŸ“ˆ EXHAUSTIVIDAD (RECALL): {recall:.4f}")
            print("   - Capacidad de detectar todos los Perros reales")

            # ğŸ¯ REQUERIMIENTO 6: PuntuaciÃ³n F1
            from sklearn.metrics import f1_score
            f1 = f1_score(y_true, y_pred)
            print(f"\nâš–ï¸  PUNTUACIÃ“N F1: {f1:.4f}")
            print("   - Balance entre PrecisiÃ³n y Recall")

            # Reporte completo de clasificaciÃ³n
            from sklearn.metrics import classification_report
            print(f"\nğŸ“„ REPORTE DE CLASIFICACIÃ“N COMPLETO:")
            print(classification_report(y_true, y_pred,
                                      target_names=['Gatos', 'Perros'],
                                      digits=4))

            # Resumen ejecutivo
            print("\n" + "="*50)
            print("ğŸ“Š RESUMEN EJECUTIVO")
            print("="*50)
            print(f"ğŸ¯ Accuracy:    {accuracy*100:.2f}%")
            print(f"ğŸ¯ Precision:   {precision*100:.2f}%")
            print(f"ğŸ“ˆ Recall:      {recall*100:.2f}%")
            print(f"âš–ï¸  F1-Score:    {f1*100:.2f}%")

            return {
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'f1_score': f1,
                'confusion_matrix': cm,
                'y_true': y_true,
                'y_pred': y_pred
            }

        except Exception as e:
            print(f"âŒ Error calculando mÃ©tricas: {e}")
            import traceback
            traceback.print_exc()
            return None

    def evaluate_model(self, validation_generator, history):
        """EvaluaciÃ³n completa del modelo"""
        print("\nğŸ“Š EVALUACIÃ“N FINAL DEL MODELO")
        print("=" * 50)

        evaluation = self.model.evaluate(validation_generator, verbose=0)
        metrics = ['loss', 'accuracy', 'precision', 'recall']

        for i, metric in enumerate(metrics):
            if i < len(evaluation):
                print(f"{metric.capitalize()}: {evaluation[i]:.4f}")

        # ğŸ†• NUEVO: EvaluaciÃ³n formal de mÃ©tricas de clasificaciÃ³n
        self.evaluate_classification_metrics(validation_generator)

        self.confusion_analysis(validation_generator)
        self.extended_validation_test(validation_generator)

    def confusion_analysis(self, validation_generator):
        """AnÃ¡lisis de matriz de confusiÃ³n"""
        print("\nğŸ” AnÃ¡lisis de Errores:")

        try:
            images, true_labels = next(validation_generator)
            predictions = self.model.predict(images, verbose=0)

            correct = 0
            cat_errors = 0
            dog_errors = 0

            for i in range(len(images)):
                pred = predictions[i][0]
                true = true_labels[i]

                pred_label = "PERRO" if pred > 0.5 else "GATO"
                true_label = "PERRO" if true > 0.5 else "GATO"

                if pred_label == true_label:
                    correct += 1
                else:
                    if true_label == "GATO":
                        cat_errors += 1
                    else:
                        dog_errors += 1

            total = len(images)
            accuracy = correct / total

            print(f"âœ… Correctas: {correct}/{total} ({accuracy:.1%})")
            print(f"âŒ Gatos mal clasificados: {cat_errors}")
            print(f"âŒ Perros mal clasificados: {dog_errors}")

        except Exception as e:
            print(f"âš ï¸  Error en anÃ¡lisis: {e}")

    def extended_validation_test(self, validation_generator):
        """Test de validaciÃ³n extendido"""
        print("\nğŸ§ª Test de ValidaciÃ³n Extendido:")

        try:
            correct_predictions = 0
            total_predictions = 0
            confidence_threshold = 0.8

            for i in range(3):
                images, labels = next(validation_generator)
                predictions = self.model.predict(images, verbose=0)

                for j in range(len(images)):
                    pred = predictions[j][0]
                    true = labels[j]

                    confidence = pred if pred > 0.5 else 1 - pred

                    if (pred > 0.5 and true > 0.5) or (pred < 0.5 and true < 0.5):
                        correct_predictions += 1

                    total_predictions += 1

            accuracy = correct_predictions / total_predictions
            print(f"ğŸ“Š PrecisiÃ³n en test extendido: {accuracy:.1%}")
            print(f"ğŸ“ˆ Muestras evaluadas: {total_predictions}")

        except Exception as e:
            print(f"âš ï¸  Error en test extendido: {e}")

    def predict_with_high_confidence(self, image_path):
        """PredicciÃ³n con anÃ¡lisis de confianza mejorado"""
        if self.model is None:
            model_files = ['best_dog_cat_model_advanced.h5', 'dog_cat_classifier_advanced.h5']
            for model_file in model_files:
                if os.path.exists(model_file):
                    print(f"ğŸ” Cargando modelo: {model_file}")
                    self.model = tf.keras.models.load_model(model_file)
                    break
            else:
                print("âŒ No hay modelo entrenado disponible.")
                return None, None, None

        try:
            img = tf.keras.preprocessing.image.load_img(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
            img_array = tf.keras.preprocessing.image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array /= 255.0

            prediction = self.model.predict(img_array, verbose=0)[0][0]
            confidence = prediction if prediction > 0.5 else 1 - prediction

            if confidence > 0.90:
                result = "PERRO ğŸ¶" if prediction > 0.5 else "GATO ğŸ±"
                confidence_level = "MUY ALTA"
            elif confidence > 0.80:
                result = "Probablemente PERRO ğŸ¶" if prediction > 0.5 else "Probablemente GATO ğŸ±"
                confidence_level = "ALTA"
            elif confidence > 0.65:
                result = "Posiblemente PERRO ğŸ¶?" if prediction > 0.5 else "Posiblemente GATO ğŸ±?"
                confidence_level = "MEDIA"
            else:
                result = "NO SEGURO ğŸ¤”"
                confidence_level = "BAJA"

            return result, confidence * 100, confidence_level

        except Exception as e:
            print(f"âŒ Error al procesar la imagen: {e}")
            return None, None, None

    def plot_detailed_training_history(self, history):
        """Grafica la evoluciÃ³n del entrenamiento de manera detallada"""
        if history is None:
            print("ğŸ“Š No hay historial de entrenamiento para graficar")
            return

        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

        ax1.plot(history.history['accuracy'], 'b-', label='Entrenamiento', linewidth=2)
        ax1.plot(history.history['val_accuracy'], 'r-', label='ValidaciÃ³n', linewidth=2)
        ax1.set_title('PrecisiÃ³n del Modelo', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Ã‰poca')
        ax1.set_ylabel('PrecisiÃ³n')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        ax2.plot(history.history['loss'], 'b-', label='Entrenamiento', linewidth=2)
        ax2.plot(history.history['val_loss'], 'r-', label='ValidaciÃ³n', linewidth=2)
        ax2.set_title('PÃ©rdida del Modelo', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Ã‰poca')
        ax2.set_ylabel('PÃ©rdida')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        if 'precision' in history.history and 'recall' in history.history:
            ax3.plot(history.history['precision'], 'g-', label='PrecisiÃ³n', linewidth=2)
            ax3.plot(history.history['recall'], 'orange', label='Recall', linewidth=2)
            ax3.set_title('PrecisiÃ³n vs Recall', fontsize=14, fontweight='bold')
            ax3.set_xlabel('Ã‰poca')
            ax3.legend()
            ax3.grid(True, alpha=0.3)

        if 'lr' in history.history:
            ax4.plot(history.history['lr'], 'purple', label='Learning Rate', linewidth=2)
            ax4.set_title('Learning Rate', fontsize=14, fontweight='bold')
            ax4.set_xlabel('Ã‰poca')
            ax4.set_ylabel('LR')
            ax4.set_yscale('log')
            ax4.legend()
            ax4.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def upload_and_predict_advanced(self):
        """Permite subir una imagen y predecir con anÃ¡lisis avanzado"""
        try:
            from google.colab import files
            print("\nğŸ“¤ Sube una imagen de un perro o gato...")

            uploaded = files.upload()

            if not uploaded:
                print("âŒ No se subiÃ³ ningÃºn archivo.")
                return

            uploaded_filename = list(uploaded.keys())[0]
            print(f"âœ… Imagen subida: {uploaded_filename}")

            result, confidence, confidence_level = self.predict_with_high_confidence(uploaded_filename)

            if result:
                print(f"\n{'='*60}")
                print("ğŸ¯ RESULTADO DE CLASIFICACIÃ“N AVANZADA")
                print(f"{'='*60}")
                print(f"ğŸ“ Imagen: {uploaded_filename}")
                print(f"ğŸ” Resultado: {result}")
                print(f"ğŸ“Š Confianza: {confidence:.2f}%")
                print(f"ğŸšï¸  Nivel de confianza: {confidence_level}")

                if confidence_level == "BAJA":
                    print("ğŸ’¡ RecomendaciÃ³n: La imagen podrÃ­a ser ambigua o de baja calidad")
                elif confidence_level == "MEDIA":
                    print("ğŸ’¡ RecomendaciÃ³n: Resultado aceptable, pero con margen de error")
                else:
                    print("ğŸ’¡ RecomendaciÃ³n: Resultado muy confiable")

                print(f"{'='*60}")

                img = Image.open(uploaded_filename)
                plt.figure(figsize=(10, 8))
                plt.imshow(img)
                plt.title(f"PredicciÃ³n: {result}\nConfianza: {confidence:.2f}% ({confidence_level})",
                         fontsize=16, fontweight='bold', pad=20)
                plt.axis('off')
                plt.tight_layout()
                plt.show()

            else:
                print("âŒ No se pudo realizar la clasificaciÃ³n.")

        except ImportError:
            print("âŒ Esta funciÃ³n solo estÃ¡ disponible en Google Colab")
        except Exception as e:
            print(f"âŒ Error: {e}")

    def test_model(self):
        """FunciÃ³n para probar con imÃ¡genes existentes"""
        print("\nğŸ” Buscando imÃ¡genes de prueba...")

        test_images = []
        for root, dirs, files in os.walk(self.data_path):
            for file in files:
                if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                    test_images.append(os.path.join(root, file))
                    if len(test_images) >= 5:
                        break
            if len(test_images) >= 5:
                break

        if test_images:
            print(f"ğŸ§ª Probando con {len(test_images)} imÃ¡genes del dataset...")
            for img_path in test_images[:3]:
                result, confidence, confidence_level = self.predict_with_high_confidence(img_path)
                if result:
                    print(f"\nğŸ“Š {os.path.basename(img_path)}: {result} ({confidence:.1f}% - {confidence_level})")
        else:
            print("âŒ No se encontraron imÃ¡genes de prueba")


def main():
    print("ğŸ• CLASIFICADOR AVANZADO DE PERROS Y GATOS ğŸˆ")
    print(f"TensorFlow version: {tf.__version__}")
    print("=" * 60)
    print("ğŸ¯ CARACTERÃSTICAS MEJORADAS:")
    print("â€¢ Transfer Learning con MobileNetV2")
    print("â€¢ Entrenamiento en dos etapas")
    print("â€¢ Data augmentation optimizado")
    print("â€¢ AnÃ¡lisis de confianza detallado")
    print("â€¢ âœ… MÃ‰TRICAS FORMALES DE CLASIFICACIÃ“N")
    print("=" * 60)

    try:
        classifier = DogCatClassifier()

        print("\nğŸ¤– Â¿QuÃ© tipo de modelo prefieres?")
        print("1. ğŸš€ Transfer Learning (Recomendado - MÃ¡s preciso)")
        print("2. ğŸ§  CNN Personalizado (MÃ¡s rÃ¡pido)")

        choice = input("Selecciona (1-2, Enter para 1): ").strip()
        if choice == "2":
            classifier.use_transfer_learning = False
            print("âœ… Usando CNN Personalizado")
        else:
            print("âœ… Usando Transfer Learning (Recomendado)")

        history = classifier.train_advanced()

        if history:
            classifier.plot_detailed_training_history(history)

            while True:
                print(f"\n{'='*50}")
                print("Â¿QuÃ© deseas hacer ahora?")
                print("1. ğŸ“¤ Subir imagen y clasificar")
                print("2. ğŸ“ Probar con imÃ¡genes existentes")
                print("3. ğŸ“Š Ver mÃ©tricas formales de clasificaciÃ³n")
                print("4. ğŸ”„ Reentrenar con diferente configuraciÃ³n")
                print("5. ğŸšª Salir")

                choice = input("Selecciona una opciÃ³n (1-5): ").strip()

                if choice == '1':
                    classifier.upload_and_predict_advanced()
                elif choice == '2':
                    classifier.test_model()
                elif choice == '3':
                    # ğŸ†• NUEVA OPCIÃ“N: MÃ©tricas formales
                    print("\nğŸ“Š Generando reporte formal de mÃ©tricas...")
                    classifier.evaluate_classification_metrics()
                elif choice == '4':
                    classifier.use_transfer_learning = not classifier.use_transfer_learning
                    model_type = "Transfer Learning" if classifier.use_transfer_learning else "CNN Personalizado"
                    print(f"ğŸ”„ Reentrenando con {model_type}...")
                    history = classifier.train_advanced()
                    if history:
                        classifier.plot_detailed_training_history(history)
                elif choice == '5':
                    print("ğŸ‘‹ Â¡Hasta luego!")
                    break
                else:
                    print("âŒ OpciÃ³n invÃ¡lida.")
        else:
            print("âŒ Error en el entrenamiento")

    except Exception as e:
        print(f"âŒ Error: {e}")
        import traceback
        traceback.print_exc()


if __name__ == "__main__":
    main()
```
